```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

# Data


```python
data = pd.read_csv('cost-revenue-clean.csv')
data
```

```python
data.describe()
```

# Visualize

```python
plt.figure(figsize=(10, 6))
plt.scatter(data.production_budget_usd, data.worldwide_gross_usd, alpha=0.4)
plt.xlabel('Бюджет')
plt.ylabel('Кассовый сбор')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
```

# Model

```python
model = LinearRegression()
```

```python
X = pd.DataFrame(data.production_budget_usd)
Y = pd.DataFrame(data.worldwide_gross_usd)
```

```python
model.fit(X.values, Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data.production_budget_usd, data.worldwide_gross_usd, alpha=0.4)

plt.plot(X.values, model.predict(X.values), color='red', linewidth=3)

plt.xlabel('Бюджет')
plt.ylabel('Кассовый сбор')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
```

```python
model.score(X.values, Y)
```
